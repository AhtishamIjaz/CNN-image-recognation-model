```python
import tensorflow
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPooling2D, Flatten, Dropout
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping

# Define CNN with Dropout
cnn = Sequential()
cnn.add(Conv2D(32, (3, 3), input_shape=(64, 64, 3), activation="relu"))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
cnn.add(Dropout(0.25))  # Drop 25% of nodes

cnn.add(Conv2D(16, (3, 3), activation="relu"))
cnn.add(MaxPooling2D(pool_size=(2, 2)))
cnn.add(Dropout(0.25))

cnn.add(Flatten())
cnn.add(Dense(128, activation='relu'))
cnn.add(Dropout(0.5))  # Drop 50% before output
cnn.add(Dense(10, activation='softmax'))  # 10 classes

# Compile the model
cnn.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Image augmentation
train_datagen = ImageDataGenerator(
    rescale=1. / 255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1. / 255)

# Training data
train_generator = train_datagen.flow_from_directory(
    r"C:\Users\Ahtisham Ijaz\Downloads\tomato\train",
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical'
)

# Validation data
test_generator = test_datagen.flow_from_directory(
    r"C:\Users\Ahtisham Ijaz\Downloads\tomato\val",
    target_size=(64, 64),
    batch_size=32,
    class_mode='categorical'
)

# Add EarlyStopping
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=3,
    restore_best_weights=True
)

# Train model
cnn.fit(
    train_generator,
    steps_per_epoch=len(train_generator),
    epochs=20,
    validation_data=test_generator,
    validation_steps=len(test_generator),
    callbacks=[early_stop]
)
```
